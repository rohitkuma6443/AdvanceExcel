## Concise Summary of Seaborn Outline

*   **What is Seaborn?** A Python data visualization library based on Matplotlib for creating informative statistical graphics.
*   **Why Use Seaborn?** Offers a high-level interface, attractive default styles, and strong integration with statistical functions and pandas DataFrames, simplifying complex plots.
*   **Difference Between Matplotlib and Seaborn:** Matplotlib is lower-level and general-purpose; Seaborn is higher-level, specialized for statistical plotting, and built on Matplotlib.
*   **Installing Seaborn:** Typically installed via `pip` or `conda` (often pre-installed in Colab).
*   **Importing Libraries:** Standard practice is to import `matplotlib.pyplot` as `plt` and `seaborn` as `sns`.
*   **Loading Built-in Datasets:** Seaborn provides functions like `sns.load_dataset()` to easily load example datasets for practice.
*   **Understanding Dataset Structure:** Involves examining the first few rows (`.head()`), data types (`.info()`), and descriptive statistics (`.describe()`) to grasp data characteristics.

## Module 2: Seaborn Basics

### Setting Themes and Styles

Seaborn allows you to quickly change the aesthetic style of your plots using `sns.set_theme()` or `sns.set_style()`. Themes control elements like font size, line thickness, and legend placement, while styles primarily affect the background and grid. Common styles include `darkgrid`, `whitegrid`, `dark`, `white`, and `ticks`.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Apply a default Seaborn theme
sns.set_theme()
print("Default Seaborn theme applied.")

# Example of setting a specific style
sns.set_style('whitegrid')
print("Seaborn style set to 'whitegrid'.")

# Let's create a simple plot to see the effect
plt.figure(figsize=(8, 5))
sns.histplot(data=tips_df, x='total_bill', kde=True)
plt.title('Total Bill Distribution with Whitegrid Style')
plt.show()
```

### Figure Size and Background

While Seaborn sets many aspects of plot aesthetics, Matplotlib functions are often used to control the overall figure size and specific background properties. `plt.figure(figsize=(width, height))` is commonly used to adjust the plot dimensions.

```python
# Reset to default style for demonstration
sns.set_theme()

# Create a figure with a specific size
plt.figure(figsize=(10, 6))

# Plotting data
sns.scatterplot(data=tips_df, x='total_bill', y='tip', hue='time')
plt.title('Total Bill vs Tip by Time (Custom Figure Size)')
plt.show()

print("Figure with custom size created and displayed.")
```

### Color Palettes

Seaborn offers a rich selection of color palettes to enhance the visual appeal and interpretability of your plots. You can choose from qualitative, sequential, and diverging palettes using `sns.color_palette()` or by specifying a `palette` argument in plotting functions.

```python
# Display some color palettes
print("Default color palette:")
sns.palplot(sns.color_palette())
plt.title('Default Palette')
plt.show()

print("\n'viridis' sequential color palette:")
sns.palplot(sns.color_palette('viridis'))
plt.title('Viridis Palette')
plt.show()

# Use a custom palette in a plot
plt.figure(figsize=(8, 5))
sns.boxplot(data=tips_df, x='day', y='total_bill', palette='pastel')
plt.title('Total Bill by Day (Pastel Palette)')
plt.show()

print("Plots with different color palettes displayed.")
```

### Titles and Labels

Clear titles and labels are essential for making your plots understandable. While some Seaborn functions can add titles and labels, it's often more flexible to use Matplotlib's `plt.title()`, `plt.xlabel()`, and `plt.ylabel()` for precise control.

```python
plt.figure(figsize=(8, 5))
sns.violinplot(data=tips_df, x='time', y='tip', hue='smoker')

# Add custom title and labels
plt.title('Tip Distribution by Time and Smoker Status', fontsize=16)
plt.xlabel('Meal Time', fontsize=12)
plt.ylabel('Tip Amount ($)', fontsize=12)
plt.show()

print("Plot with custom title and labels displayed.")
```

### Grid and Axis Customization

Customizing the grid and axis allows for better readability and focus on specific data points. Matplotlib functions like `plt.grid()`, `plt.xticks()`, `plt.yticks()`, `plt.xlim()`, and `plt.ylim()` are used for this purpose.

```python
plt.figure(figsize=(9, 6))
sns.histplot(data=tips_df, x='total_bill', bins=15, color='skyblue')

# Customize grid and axis
plt.grid(True, linestyle='--', alpha=0.6) # Add a dashed grid
plt.xticks(rotation=45) # Rotate x-axis labels
plt.xlim(0, 50) # Set x-axis limits
plt.xlabel('Total Bill Amount', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('Total Bill Distribution with Custom Grid and Axis', fontsize=16)
plt.show()

print("Plot with custom grid and axis displayed.")
```

### Saving Charts

After creating a chart, you can save it to a file in various formats (e.g., PNG, JPEG, PDF) using Matplotlib's `plt.savefig()`. This is crucial for reports, presentations, or sharing your visualizations.

```python
# Create a simple plot
plt.figure(figsize=(7, 4))
sns.countplot(data=tips_df, x='day', palette='viridis')
plt.title('Number of Customers by Day')

# Save the chart to a file
file_name = 'customers_by_day.png'
plt.savefig(file_name, dpi=300, bbox_inches='tight')
plt.show()

print(f"Chart saved as '{file_name}'.")
```

## Module 3: Relational Plots

### Line Plot

`sns.lineplot()` is used to show the relationship between two numerical variables, especially when one of them represents time or sequence. It is particularly useful for visualizing trends over time.

```python
# To demonstrate a line plot, let's create a hypothetical time-series dataset first
import numpy as np

# Create a dummy time series data
dates = pd.to_datetime(pd.date_range(start='2023-01-01', periods=len(tips_df)))
dummy_data = pd.DataFrame({
    'date': dates,
    'total_bill': tips_df['total_bill'].rolling(window=7).mean().fillna(tips_df['total_bill']).values,
    'tip': tips_df['tip'].rolling(window=7).mean().fillna(tips_df['tip']).values
})

plt.figure(figsize=(10, 6))
sns.lineplot(data=dummy_data, x='date', y='total_bill')
plt.title('Daily Total Bill (Simulated Time Series)')
plt.xlabel('Date')
plt.ylabel('Total Bill')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

print("Simple line plot displayed showing 'total_bill' over 'date'.")
```
